```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
header = {
    'User-Agent': 'your user agent via whats myuseragent website' }
```

```python
def get_reviews(base_url, num_pages):
    data = []
    data_list = []
    for j in range(1, num_pages + 1):
        url = f'{base_url}{j}/'
        print(url)
        response = requests.get(url=url, headers=header).content
        soup = BeautifulSoup(response, 'html.parser')
        reviews = soup.find_all('article', {'itemprop': 'review'})
        tables = soup.find_all('table', class_='review-ratings')

        for review, table in zip(reviews, tables):
            rows = table.find_all('tr')
            row_data = {}

            for row in rows:
                cols = row.find_all('td')
                if len(cols) == 2:
                    topic = cols[0].text.strip()
                    
                    star_elements = cols[1].find_all('span', class_='star fill')
                    if star_elements:
                        value = len(star_elements)
                    else:
                        value = cols[1].text.strip()

                    row_data[topic] = value

            data_list.append(row_data)

            name = review.find('span', {'itemprop': 'name'}).text
            date = review.find('time', {'itemprop': 'datePublished'}).text
            type_of_review = [review.find('div', {'itemprop': 'reviewBody'}).text.strip()]
            typeofreview = []
            comments=[]
            for i in type_of_review:
                if i[0:1].isalpha():
                    if i[0:12]=='Not Verified':
                        typeofreview.append(i[0:12])
                        comments.append(i[15:])
                    else:
                        typeofreview = None
                        comments.append(i)
                        
                elif i[0:1].isdigit():
                    typeofreview = None
                    comments.append(i)
                else:
                    if i[2:15]=='Trip Verified':
                        typeofreview.append(i[2:15])
                        comments.append(i[17:])
                    else:
                        typeofreview.append(i[2:17])
                        comments.append(i[20:])
                    
            #comments = list(i[15:] for i in type_of_review)
            #comments=review.find('div', {'itemprop': 'reviewBody'}).text.strip()
            if review.find('span', {'itemprop': 'ratingValue'}) is not None:
                ratings = review.find('span', {'itemprop': 'ratingValue'}).text
            else:
                ratings = None
            heading = review.find('h2', {'class': 'text_header'}).text
            temp = [name, date, typeofreview, ratings, comments]
            data.append(temp)

        print('Scraped Page', j)

    # Ensure that data and data_list are of the same length
    # If data_list has fewer entries, append empty dictionaries
    while len(data_list) < len(data):
        data_list.append({})

    # Define the topics and create the DataFrame for table data
    topics = [
        'Aircraft', 'Type Of Traveller', 'Seat Type', 'Route', 'Date Flown',
        'Seat Comfort', 'Cabin Staff Service', 'Food & Beverages',
        'Inflight Entertainment', 'Ground Service', 'Wifi & Connectivity',
        'Value For Money', 'Recommended'
    ]
    df1 = pd.DataFrame(data_list)

    for topic in topics:
        if topic not in df1.columns:
            df1[topic] = None

    df1 = df1[topics]

    indexes = ['Name', 'Date', 'Type_of_review', 'Rating', 'Review']
    df2 = pd.DataFrame(data, columns=indexes)

    df = pd.concat([df2, df1], axis=1)

    return df
```

```python
base_url='https://www.airlinequality.com/airline-reviews/air-india/page/'
df=get_reviews(base_url,136)
df.to_excel('airline_reviews_dataset.xlsx',index=False)
```
